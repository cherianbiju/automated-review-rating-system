# Import required library import pandas as pd

```python
import pandas as pd
```

# Load the raw dataset

```python
data_path = "../data/dataset/Amazon_Product_Review/Reviews.csv" 
df = pd.read_csv(data_path)
```

```python
df.head(10)
```

# Keep only review and rating 

```python
df = df[['Score', 'Text']].dropna()
df.sample(10)
```

# STANDARDIZE REVIEW TEXT (REMOVE UNWANTED SPACES AND CHARACTERS)

```python
df['Text'] = df['Text'].str.strip()
```

```python
df['Text_clean'] = df['Text'].str.lower()

import re
df['Text_clean'] = df['Text_clean'].str.replace(r'[^\w\s]', '', regex=True)

df.duplicated(subset='Text_clean')
```

# CHECKING FOR DUPLICATES

```python
df.duplicated(subset='Text_clean').sum()
```

```python
duplicates = df[df.duplicated(subset='Text', keep=False)]
duplicates.sample(10, random_state=42)
```

```python
df.head(10)
```

# DROPPING DUPLICATES

```python
df = df.drop_duplicates(subset='Text_clean', keep='first')
```

```python
df.duplicated().sum()
```

# CHECKING AND REMOVING EMPTY REVIEWS

```python
empty_count = (df['Text_clean'] == '').sum()
print(empty_count)
```

```python
df = df[df['Text_clean'] != '']
```

```python
empty_count = (df['Text_clean'] == '').sum()
print(empty_count)
```

# WRAPPING THE DATA CLEANING PROCEDURE

```python
df.head(10)
```

```python
df = df.drop(columns=['Text'])
```

```python
df.sample(10)
```

```python
df = df.rename(columns={'Text_clean': 'Reviews','Score': 'Rating'})
```

```python
df.head(5)
```

```python
# Count how many times each rating appears

rating_counts = df['Rating'].value_counts().sort_index()

print(rating_counts)
```

# SAVING CLEANED DATASET TO CSV FILE

```python
df.to_csv("../data/cleaned_dataset/amazon_cleanedreview.csv", index=False)
```
