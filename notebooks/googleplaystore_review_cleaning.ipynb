# Import required library import pandas as pd

```python
import pandas as pd
```

# Load the raw dataset

```python
data_path = "../data/dataset/Google_playstore_Review/reviews.csv" 
df = pd.read_csv(data_path)
```

```python
df.head(10)
```

# Keep only 'content' (review) and 'score' (rating) 

```python
df = df[['content', 'score']].dropna()
df.sample(10)
```

# STANDARDIZE REVIEW TEXT (REMOVE UNWANTED SPACES AND CHARACTERS)

```python
df['content'] = df['content'].str.strip()
```

```python
df['content_clean'] = df['content'].str.lower()

import re
df['content_clean'] = df['content_clean'].str.replace(r'[^\w\s]', '', regex=True)

df.duplicated(subset='content_clean')
```

# CHECKING FOR DUPLICATES

```python
df.duplicated(subset='content_clean').sum()
```

```python
duplicates = df[df.duplicated(subset='content', keep=False)]
duplicates.sample(10, random_state=42)
```

```python
df.head(10)
```

# DROPPING DUPLICATES

```python
df = df.drop_duplicates(subset='content_clean', keep='first')
```

```python
df.duplicated().sum()
```

# CHECKING AND REMOVING EMPTY REVIEWS

```python
empty_count = (df['content_clean'] == '').sum()
print(empty_count)
```

```python
df = df[df['content_clean'] != '']
```

```python
empty_count = (df['content_clean'] == '').sum()
print(empty_count)
```

# WRAPPING THE DATA CLEANING PROCEDURE

```python
df.head(10)
```

```python
df = df.drop(columns=['content'])
```

```python
df.sample(10)
```

```python
df = df.rename(columns={'content_clean': 'Reviews','score': 'Rating'})
```

```python
df.head(5)
```

```python
# Count how many times each rating appears

rating_counts = df['Rating'].value_counts().sort_index()

print(rating_counts)
```

# SAVING CLEANED DATASET TO CSV FILE

```python
df.to_csv("../data/cleaned_dataset/google_playstore_cleanedreview.csv", index=False)
```
