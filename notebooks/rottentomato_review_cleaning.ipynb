# IMPORT LIBRARIES AND DATASET

```python
import pandas as pd
import numpy as np
```

```python
data_path="../data/dataset/Rotten_Tomatoes_Review/rotten_tomatoes_critic_reviews.csv"
df=pd.read_csv(data_path)
```

```python
df.head(20)
```

```python
df=df[['review_score','review_content']]
```

```python
df.head()
```

```python
df=df.dropna(subset='review_score').reset_index(drop=True)
```

```python
df.head()
```

# HANDLING FRACTIONAL RATINGS AND RATING USING LETTERS

```python
grade_map = {'A':5, 'B':4, 'C':3, 'D':2, 'F':1}
df['Rating'] = df['review_score'].map(grade_map)

# 2️⃣ Handle fractions like 3.5/5 or 3/5
mask = df['Rating'].isna() & df['review_score'].notna()   # rows not yet mapped
left = df.loc[mask, 'review_score'].str.split('/', expand=True)[0]  # part before '/'

# convert to float safely and take integer part (floor)
df.loc[mask, 'Rating'] = pd.to_numeric(left, errors='coerce').fillna(0).astype(int)
```
